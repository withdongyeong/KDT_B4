```python
import sys
import torchvision.transforms
```

```python
from tqdm import tqdm
from pathlib import Path
import matplotlib.pyplot as plt

# attributes 파일 생성
f = open(data_path + os.path.sep + "attributes.txt", "w")
attribute_list = ["예술적인", "형식적인", "우아한", "현대적인", "날카로운", "부드러운", "굵직한", "가독성있는", "필기체", "매력적인"]
f.write("예술적인 형식적인 우아한 현대적인 날카로운 부드러운 굵직한 가독성있는 필기체 매력적인\n")
# fot att2font
for file_name in tqdm(file_list):
    temp_path = img_path + os.path.sep + Path(file_name).stem.replace(" ", "")
    if not os.path.isdir(temp_path):
        os.mkdir(temp_path)
    # random attributes
#     attributes = "  ".join([str(round(random.uniform(0, 1), 3)) for x in range(10)])
    # set attribute custom
    # 배경 이미지 크기
    W, H = (150*40, 300)
    # 흰색 배경
    image = Image.new('RGB', (W, H), (255, 255, 255))
    # 폰트 설정
    font_size = 150
    font = ImageFont.truetype(ttf_dir + os.path.sep + file_name, font_size)
    # 이미지 생성
    draw = ImageDraw.Draw(image)
    # 글자 크기
    w, h = draw.textsize("다", font=font)
    # 배경 중간에 글자 배치
    draw.text((0, (H - h) / 2), "값같곬곶깎넋늪닫닭닻됩뗌략몃밟볘뺐뽈솩쐐앉않얘얾엌옳읊죡쮜춰츄퀭틔핀핥후", fill="black", font=font)
    plt.figure(figsize=(11, 11))
    plt.imshow(image)
    plt.show()
    plt.imshow(image)
    plt.show()
    #image = stretch(image, 2)
#     image = perspective(image)
    print(file_name)
    input_list = []
    for item in attribute_list:
        attr = input(item + ":")
        input_list.append(attr)
    try:
        attributes = "  ".join([str(round(float(x) / 10, 1)) for x in input_list])
    except:
        print("한번 더 기회를 주겠어")
        input_list = []
        for item in attribute_list:
            attr = input(item + ":")
            input_list.append(attr)
            attributes = "  ".join([str(round(float(x) / 10, 1)) for x in input_list])
    print("attributes set : ", attributes)
    
    
    for i, character in tqdm(enumerate(kor_characters)):
        if i < 10:
            continue
        # 배경 이미지 크기
        idx = '{0:02d}'.format(i)
        W, H = (256, 256)
        # 흰색 배경
        image = Image.new('RGB', (W, H), (255, 255, 255))
        # 폰트 설정
        font_size = 130
        font = ImageFont.truetype(ttf_dir + os.path.sep + file_name, font_size)
        # 이미지 생성
        draw = ImageDraw.Draw(image)
        # 글자 크기
        w, h = draw.textsize(character, font=font)
        
        # 배경 중간에 글자 배치
        draw.text(((W - w ) / 2, (H - h) / 2), character, fill="black", font=font)
        image.save(img_path + os.path.sep + Path(file_name).stem.replace(" ", "") + os.path.sep + idx +".png")
        f.write(Path(file_name).stem.replace(" ", "") + "/" + idx +".png" + "  " + attributes + "\n")        
        
f.close()

# for dmfont

# todo
```
