```python
data_root = "data"
if not os.path.isdir(data_root):
    os.mkdir(data_root)

data_set_name = "explor_all"
data_path = data_root + os.path.sep + data_set_name
if not os.path.isdir(data_path):
    os.mkdir(data_path)

img_path = data_path + os.path.sep + "image"
if not os.path.isdir(img_path):
    os.mkdir(img_path)
```

```python
import random
attribute_template = "  ".join([str(round(random.uniform(0, 1), 2)) for x in range(37)])
print(attribute_template)
print(type(attribute_template))
```

```python
from tqdm import tqdm
from pathlib import Path

# attributes 파일 생성
f = open(data_path + os.path.sep + "attributes.txt", "w")
f.write("angular artistic attention-grabbing attractive bad boring calm capitals charming clumsy complex cursive delicate disorderly display dramatic formal fresh friendly gentle graceful happy italic legible modern monospace playful pretentious serif sharp sloppy soft strong technical thin warm wide\n")

# fot att2font
for file_name in tqdm(file_list):
    temp_path = img_path + os.path.sep + Path(file_name).stem.replace(" ", "")
    if not os.path.isdir(temp_path):
        os.mkdir(temp_path)
    # random attribute for each font
    attributes = "  ".join([str(round(random.uniform(0, 1), 3)) for x in range(37)])
    for i, character in tqdm(enumerate(kor_characters)):
        if i < 10:
            continue
        # 배경 이미지 크기
        idx = '{0:02d}'.format(i)
        W, H = (256, 256)
        # 흰색 배경
        image = Image.new('RGB', (W, H), (255, 255, 255))
        # 폰트 설정
        font_size = 130
        font = ImageFont.truetype(ttf_dir + os.path.sep + file_name, font_size)
        # 이미지 생성
        draw = ImageDraw.Draw(image)
        # 글자 크기
        w, h = draw.textsize(character, font=font)
        # 배경 중간에 글자 배치
        draw.text(((W - w ) / 2, (H - h) / 2), character, fill="black", font=font)
        image.save(img_path + os.path.sep + Path(file_name).stem.replace(" ", "") + os.path.sep + idx +".png")
#         image.save(img_path + os.path.sep + Path(file_name).stem.replace(" ", "") + os.path.sep + character +".png")
        f.write(Path(file_name).stem.replace(" ", "") + "/" + idx +".png" + "  " + attributes + "\n")
#         f.write(Path(file_name).stem.replace(" ", "") + "/" + character +".png" + "  " + attribute_template + "\n")
        
f.close()

# for dmfont

# todo
```
